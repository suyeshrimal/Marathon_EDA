# Data Processing

```python
df2.head(5)
```

```python
#calculating average speed by event distance and gender
df2.groupby(['Event distance/length','Athlete gender'])['Athlete average speed'].mean()
```

```python
#Calculating Seasons for the event along with the average speed in a particular season and total races in that season
```

```python
df2['event month']=df2['Event dates'].str.split('.').str.get(1).astype('int')
```

```python
df2['event season']=df2['event month'].apply(lambda x: 'winter' if x > 11 else 'fall' if x >8 else 'summer' if x > 5 else 'spring' if x>2 else 'winter' ) 
```

```python
df2.head(10)
```

```python
df2.groupby('event season')['Athlete average speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```
